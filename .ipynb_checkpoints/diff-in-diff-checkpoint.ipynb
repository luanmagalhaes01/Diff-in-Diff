# Analisando o impacto de uma campanha de cupons em pedidos de delivery: Um estudo de Diferenças em Diferenças

## Introdução

O uso de **campanhas de desconto e cupons** é uma estratégia amplamente adotada no mercado de aplicativos de delivery, visando aumentar a retenção de clientes e impulsionar o número de pedidos. No entanto, **medir o impacto real dessas ações promocionais** sobre o comportamento dos consumidores pode ser um desafio. Uma maneira eficiente de abordar essa questão é através da **metodologia de *Diferenças em Diferenças* (DD)**, que permite isolar o efeito da campanha de cupons ao comparar **grupos de clientes expostos e não expostos** antes e depois da ação.

Neste estudo, utilizamos dados fictícios para simular o comportamento de clientes de um aplicativo de entrega de comida. O objetivo é **entender se houve um aumento significativo no número de pedidos** entre os clientes expostos a cupons de desconto em comparação com aqueles que não receberam a oferta. Para isso, comparamos o número de pedidos realizados por ambos os grupos antes e depois da campanha, controlando por variáveis como idade, salário e localização.

A análise foi estruturada com base no método de Diferenças em Diferenças, permitindo obter uma estimativa mais precisa do impacto da campanha sobre os pedidos e oferecendo insights sobre a efetividade dessa campanha de marketing.

## Objetivo do Projeto:

Avaliar o impacto de uma campanha de cupons de desconto no número de pedidos realizados por clientes em um aplicativo de delivery.

## Problema a ser Resolvido:

- A campanha foi eficaz em aumentar o número de pedidos?
- Qual foi o impacto nos clientes expostos versus os não expostos?

## Diferenças em Diferenças (DD)

O método de diferença em diferenças compara a diferença no resultado de interesse (como vendas, desempenho, saúde) entre dois grupos: um grupo de tratamento (que foi exposto à intervenção) e um grupo de controle (que não foi exposto). Essa comparação é feita antes e depois da intervenção, permitindo identificar o efeito da mudança ao longo do tempo.

### Comparar o Antes e Depois do Programa Permite Avaliar o Impacto Causal?

- **Não!** Comparar apenas o antes e depois pode ser problemático porque muitos outros fatores podem influenciar o resultado ao longo do tempo. Por exemplo, no caso de uma análise de emprego, fatores econômicos externos, políticas governamentais, ou mudanças sazonais podem afetar o nível de emprego, dificultando a identificação do impacto isolado do programa.

### Comparar Quem Recebeu o Programa com Quem Não Recebeu Funciona?

- **Não!** Comparar diretamente quem recebeu o programa com quem não recebeu também pode gerar vieses. Isso ocorre porque podem existir variáveis não observadas que determinam por que alguns indivíduos receberam o programa enquanto outros não. Esses fatores não observados podem influenciar os resultados, confundindo a análise e levando a uma estimativa incorreta do efeito do programa.

A abordagem de Diferenças em Diferenças combina dois cenários problemáticos (comparações antes e depois, e comparações entre os que escolheram se inscrever e os que não se inscreveram) para produzir uma estimativa mais precisa do cenário contrafactual.

- O cenário contrafactual estimado é a mudança nos resultados do grupo de controle.
- Os grupos de tratamento e de controle não precisam, necessariamente, ter as mesmas condições antes da intervenção.
- Para que o método seja válido, o grupo de controle deve representar com precisão a mudança nos resultados que o grupo de tratamento teria experimentado na ausência do tratamento.

A melhor maneira de entender o conceito é visualizá-lo graficamente:

```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame com os dados fornecidos
dados = pd.DataFrame({
    'periodo': ['Antes', 'Depois', 'Antes', 'Depois'],
    'grupo': ['Controle', 'Controle', 'Tratamento', 'Tratamento'],
    'resultado': [4.6, 8.7, 2.7, 7.8],
    'letras': ['C', 'D','A', 'B']  # Adicionar letras para identificar os pontos
})

# Criar gráfico de linha
plt.figure(figsize=(8, 4))  # Aumentando o tamanho da figura
for grupo in dados['grupo'].unique():
    subset = dados[dados['grupo'] == grupo]
    plt.plot(subset['periodo'], subset['resultado'], marker='o', label=grupo, linewidth=2, markersize=8)

    # Adicionando as letras e os valores nos pontos
    for i in range(len(subset)):
        plt.text(subset['periodo'].iloc[i], subset['resultado'].iloc[i] + 0.5, 
                 f"{subset['letras'].iloc[i]} ({subset['resultado'].iloc[i]})", 
                 ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

# Adicionando título e rótulos
plt.title('Diferenças em Diferenças', fontsize=14)
plt.xlabel('Período', fontsize=12)
plt.ylabel('Resultado', fontsize=12)
plt.legend(title='Grupo')

plt.grid(False)  # Removendo o grid (linhas de fundo)
plt.xticks(subset['periodo'])  # Ajustando o layout
plt.ylim(0, 10)  # Ajustando os limites do eixo Y

plt.tight_layout(pad=2.0)  # Aumentando o espaço entre os elementos

# Exibindo o gráfico
plt.show()
```

* Na metodologia de diferenças em diferenças, podemos calcular o impacto do programa:

$$
DD = (B - A) - (D - C)
$$

* No exemplo acima, o impacto é dado pela diferença entre as duas diferenças:

$$
\begin{align*}
DD & = (B - A) - (D - C) \\
   & = (7.8 - 2.7) - (8.7 - 4.6) \\
   & = 5.1 - 4.1 \\
   & = 1
\end{align*}
$$


* Calculamos a diferença no resultado entre as situações antes e depois para o grupo de tratamento **(B - A)**.
* Calculamos a diferença no resultado entre as situações antes e depois para o grupo de controle **(D - C)**.
* Em seguida, calculamos a diferença entre a diferença nos resultados para o grupo de tratamento **(B - A)** e a diferença para o grupo de controle **(D - C)**, ou seja, **DD = (B - A) - (D - C)**.
* Esta é a "diferença em diferenças".

### Como o método DD pode ser útil?

O método de Diferenças em Diferenças (DD) permite tratar das diferenças entre os grupos de tratamento e de controle que são constantes ao longo do tempo. No entanto, ele não elimina as diferenças que mudam ao longo do tempo entre esses grupos.

Para que o DD forneça uma estimativa válida do contrafactual, é necessário assumir que não existem diferenças variáveis ao longo do tempo entre os grupos. Além disso, eventos ou mudanças externas que afetam os grupos de maneira desigual, como uma mudança de política ou uma crise econômica, podem distorcer as estimativas.

A principal limitação do método é a suposição de que, na ausência da intervenção, as tendências dos grupos (tratamento e controle) teriam seguido caminhos paralelos. Se essa suposição não for válida, as estimativas podem ser enviesadas.

## Preparação dos Dados

**Fonte de Dados:**  

A base de dados foi gerada de forma fictícia no pandas, simulando o comportamento de 2.000 clientes antes e depois da campanha. Esse cenário foi criado para possibilitar a análise do impacto da ação de marketing, comparando o efeito nos grupos expostos e não expostos ao cupom de desconto.
  

**Variáveis Incluídas:**
- `client_id`: Identificador único de cada cliente.  
- `exposto_ao_cupom`: Indica se o cliente foi exposto à campanha de cupons (1 = sim, 0 = não).  
- `pedidos_antes`: Número de pedidos antes da campanha.  
- `pedidos_apos`: Número de pedidos após a campanha.  
- Variáveis de controle: `idade`, `salario`, `localizacao`.  


```python
import numpy as np

# Definindo o número de clientes
np.random.seed(42)
n_clients = 2000

# Gerando a base de dados fictícia
df = pd.DataFrame({
    'client_id': range(1, n_clients + 1),
    'exposto_ao_cupom': np.random.choice([0, 1], size=n_clients, p=[0.5, 0.5]),  # 50% expostos e 50% não
    'idade': np.random.randint(18, 65, size=n_clients),
    'salario': np.random.normal(5000, 2000, size=n_clients).round(2),
    'localizacao': np.random.choice(['Urbana', 'Suburbana', 'Rural'], size=n_clients, p=[0.6, 0.3, 0.1]),
})
```

```python
df.head()
```

**Separação dos Períodos:**

Após a geração da base de dados, os dados foram organizados em dois períodos: antes e depois da campanha.

```python
# Gerando os pedidos antes e depois da campanha com base na exposição
# Antes da campanha (sem efeito)
df['pedidos_antes'] = np.where(df['exposto_ao_cupom'] == 1,
                               np.random.poisson(4, size=n_clients),
                               np.random.poisson(3, size=n_clients))

# Após a campanha (com efeito para os expostos)
df['pedidos_apos'] = np.where(df['exposto_ao_cupom'] == 1,
                              df['pedidos_antes'] + np.random.poisson(2, size=n_clients), # Aumento p/ expostos
                              df['pedidos_antes'] + np.random.poisson(1, size=n_clients)) # Pequeno aumento p/ não expostos

# Preparar os dados para Diferenças em Diferenças
# Dados antes da campanha
df_before = df[['client_id', 'exposto_ao_cupom', 'idade', 'salario', 'localizacao', 'pedidos_antes']].copy()
df_before['tempo'] = 0  # Antes da campanha
df_before.rename(columns={'pedidos_antes': 'pedidos'}, inplace=True)

# Dados após a campanha
df_after = df[['client_id', 'exposto_ao_cupom', 'idade', 'salario', 'localizacao', 'pedidos_apos']].copy()
df_after['tempo'] = 1  # Depois da campanha
df_after.rename(columns={'pedidos_apos': 'pedidos'}, inplace=True)

# Concatenar os dois DataFrames (antes e depois)
df_did = pd.concat([df_before, df_after])

df_did.head()
```

## Análise do Modelo de Diferenças em Diferenças

### Cálculo Manual de Diferenças em Diferenças (DD)

Para cada combinação de grupo (tratamento/controle) e tempo (antes/depois), calculamos as médias do resultado de interesse, neste caso, o número de pedidos.

```python
#  Média dos pedidos após a campanha para o grupo exposto (tratamento).
A = df_did[['pedidos']][(df_did['exposto_ao_cupom'] == 1) & (df_did['tempo'] == 1)].mean().values[0]

# Média dos pedidos antes da campanha para o grupo exposto (tratamento).
B = df_did[['pedidos']][(df_did['exposto_ao_cupom'] == 1) & (df_did['tempo'] == 0)].mean().values[0]

# Média dos pedidos após a campanha para o grupo não exposto (controle).
C = df_did[['pedidos']][(df_did['exposto_ao_cupom'] == 0) & (df_did['tempo'] == 1)].mean().values[0]

# Média dos pedidos antes da campanha para o grupo não exposto (controle).
D = df_did[['pedidos']][(df_did['exposto_ao_cupom'] == 0) & (df_did['tempo'] == 0)].mean().values[0]
```

#### Primeira Diferença (A - B):

Aqui calculamos a diferença no número de pedidos apenas para o grupo tratado (exposto ao cupom), entre o período antes e depois da intervenção.

```python
A-B
```

#### Segunda Diferença (C - D):

Agora, calculamos a diferença no número de pedidos para o grupo de controle (não exposto ao cupom) entre o período antes e depois.

```python
C-D
```

#### Diferença em Diferenças:

O passo final é subtrair a segunda diferença da primeira:

```python
DD = (A-B)-(C-D)
DD
```

O valor final de Diferença em Diferenças 1.052 reflete o efeito estimado da campanha de cupons sobre os pedidos.

Isso indica que, após controlar pelas mudanças que ocorreram no grupo de controle, o efeito da campanha de cupons foi aumentar os pedidos em aproximadamente 1.05 pedidos adicionais, em média, para aqueles que foram expostos ao cupom.

#### Teste t:

Para avaliar se as diferenças observadas entre os grupos de tratamento e controle, antes e depois da intervenção, são estatisticamente significativas, realizamos um teste t.

O teste t é uma ferramenta estatística usada para comparar a média de dois grupos e verificar se as diferenças observadas entre as médias são significativas ou podem ter ocorrido por acaso.

O que o teste t faz:
- **Hipótese nula (H₀**): As médias dos dois grupos são iguais, ou seja, não há diferença significativa entre elas.
- **Hipótese alternativa (H₁)**: As médias dos dois grupos são diferentes, ou seja, há uma diferença significativa entre elas.

- O teste calcula uma estatística chamada valor de **t** e um **p-valor**, que nos ajuda a decidir se devemos rejeitar a hipótese nula ou não.
  
  - Se **o p-valor** for menor que um certo nível de significância (geralmente 0.05 ou 5%), rejeitamos a hipótese nula e concluímos que as médias são significativamente diferentes.
  
  - Caso contrário, não há evidências suficientes para rejeitar a hipótese nula.

```python
# Filtrar os dados para os grupos expostos e não expostos
expostos_antes = df_did['pedidos'][(df_did['exposto_ao_cupom'] == 1) & (df_did['tempo'] == 0)]
expostos_depois = df_did['pedidos'][(df_did['exposto_ao_cupom'] == 1) & (df_did['tempo'] == 1)]

nao_expostos_antes = df_did['pedidos'][(df_did['exposto_ao_cupom'] == 0) & (df_did['tempo'] == 0)]
nao_expostos_depois = df_did['pedidos'][(df_did['exposto_ao_cupom'] == 0) & (df_did['tempo'] == 1)]

# Cálculo das médias
media_expostos_antes = expostos_antes.mean()
media_expostos_depois = expostos_depois.mean()

media_nao_expostos_antes = nao_expostos_antes.mean()
media_nao_expostos_depois = nao_expostos_depois.mean()

# Teste t para expostos (antes vs depois)
t_stat_expostos, p_val_expostos = stats.ttest_ind(expostos_antes, expostos_depois)

# Teste t para não expostos (antes vs depois)
t_stat_nao_expostos, p_val_nao_expostos = stats.ttest_ind(nao_expostos_antes, nao_expostos_depois)

# Resultados
print("Resultados do Teste T:")
print(f"Expostos - Média Antes: {media_expostos_antes}, Média Depois: {media_expostos_depois}")
print(f"Expostos - t-statistic: {t_stat_expostos}, p-value: {p_val_expostos}\n")

print(f"Não Expostos - Média Antes: {media_nao_expostos_antes}, Média Depois: {media_nao_expostos_depois}")
print(f"Não Expostos - t-statistic: {t_stat_nao_expostos}, p-value: {p_val_nao_expostos}")
```

O p-value é extremamente baixo para ambos os grupos (muito menor que 0.05). Isso indica que podemos rejeitar a hipótese nula com alta confiança. O p-value para o grupo exposto é particularmente baixo, o que sugere uma diferença muito significativa. ### Diferenças em Diferenças via Regressão Linear

#### Por que é melhor fazer a regressão?

A regressão captura interações mais complexas entre variáveis e oferece métricas como p-valores e intervalos de confiança, avaliando a significância dos resultados. Com isso, temos uma análise mais robusta e informativa.

#### Equação do modelo

A equação básica para a regressão de Diferenças em Diferenças (DiD) pode ser expressa da seguinte forma:

$$
Y_{it} = \alpha + \beta_1 \text{Tratamento}_i + \beta_2 \text{Tempo}_t + \beta_3 (\text{Tratamento}_i \times \text{Tempo}_t) + \epsilon_{it}
$$

Onde,

- `Yt` é o resultado de interesse.  
- `Tratamento` é a variável que indica se a unidade foi ou não exposta à intervenção.  
- `Tempo` é a variável que indica se é antes ou depois da intervenção.  

#### Interação *Tratamento × Tempo*

Para rodar a regressão linear de Diferenças em Diferenças, é necessário criar uma variável que capture a interação entre o tempo e a exposição à campanha, que será o diferencial para medir o impacto.

```python
# Criar a interação entre o tempo e a exposição (variável DiD)
df_did['tempo_exposto'] = df_did['tempo'] * df_did['exposto_ao_cupom']
```

O termo de interação `Tratamento × Tempo` captura o efeito adicional do tratamento após a intervenção, ou seja, ele mede a diferença no resultado entre o grupo de tratamento e o grupo de controle, considerando o tempo (DD).

Essa variável será o foco da análise, indicando o efeito da campanha (o quanto os pedidos variaram entre os clientes expostos antes e depois).

#### Regressão linear

A regressão linear vai ter como variável dependente o número de pedidos (`pedidos`), e as variáveis explicativas serão o tempo, a exposição à campanha e a interação entre ambos.

```python
import statsmodels.api as sm

model = sm.OLS(df_did['pedidos'], 
                sm.add_constant(df_did[['tempo', 'exposto_ao_cupom', 'tempo_exposto']])).fit()

# Exibir os resultados do modelo
print(model.summary())
```

#### Resumo dos Resultados

**R-squared (R²)**: O valor de 0,215 indica que aproximadamente 21,5% da variação no número de pedidos pode ser explicada pelas variáveis incluídas no modelo. Embora o modelo capture parte da variação, ainda há muitos fatores não observados que podem estar influenciando o número de pedidos.  


**F-statistic**: O valor de 364,3 com um p-valor de 3,91e-209 sugere que o modelo, como um todo, é estatisticamente significativo (pelo menos uma das variáveis independentes tem um efeito significativo sobre o número de pedidos).

#### Coeficientes

- **Constante (const)**: O coeficiente de 2.9818 indica que, quando todas as variáveis independentes estão em zero, espera-se que o número médio de pedidos seja aproximadamente 2.98. 

- **Tempo (tempo)**: O coeficiente de 0.9706 sugere que, em média, o número de pedidos aumenta em cerca de 0.97 unidades após a campanha, independentemente da exposição aos cupons. O p-valor associado (0.000) indica que esse efeito é estatisticamente significativo.

- **Exposto ao cupom (exposto_ao_cupom)**: O coeficiente de 0.9264 indica que os clientes expostos aos cupons, em média, fazem cerca de 0.93 pedidos a mais em comparação com aqueles que não foram expostos, no mesmo período. Assim como o tempo, esse efeito também é estatisticamente significativo (p-valor = 0.000).

- **Interação (tempo_exposto)**: O coeficiente de 1.0521 representa o efeito adicional sobre o número de pedidos para clientes que foram expostos aos cupons após a campanha. Isso sugere que, além do aumento médio nos pedidos devido ao tempo e à exposição, **os clientes expostos à campanha fazem em média 1.05 pedidos a mais após a campanha em comparação com os não expostos**. Esse coeficiente também é estatisticamente significativo (p-valor = 0.000).

## Conclusão

- Os resultados da regressão linear confirmam que a campanha de cupons teve um impacto positivo e significativo no número de pedidos.


- Embora a explicação do modelo (R²) seja moderada, a significância estatística dos coeficientes sugere que a campanha de cupons é uma estratégia eficaz para aumentar os pedidos no aplicativo de delivery analisado.
